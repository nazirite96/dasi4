```python
cc_df['trans_date_trans_time'] = pd.to_datetime(cc_df['trans_date_trans_time'])
```

```python
cc_df['hour'] = cc_df['trans_date_trans_time'].dt.hour
```

```python
def hour_func(x):
    if (x >= 6) & (x < 12):
        return 'morning'
    elif (x >= 12) & (x < 18):
        return 'afternoon'
    elif (x >= 18) & (x < 23):
        return 'night'
    else:
        return 'evening'
```

```python
cc_df['hour_cat'] = cc_df.apply(lambda x: hour_func(x['hour']),axis = 1) 
```
